```python
import tensorflow as tf
import cv2
import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator #, image
```

```python
train=ImageDataGenerator(rescale=1/255)
```

```python
#violence
train_dataset=train.flow_from_directory('/kaggle/input/ucf-crime-dataset/Train',
                                        target_size=(224, 224),
                                        batch_size=3,
                                        class_mode='binary')
#non-violence
validation_dataset=train.flow_from_directory('/kaggle/input/ucf-crime-dataset/Test',
                                        target_size=(224, 224),
                                        batch_size=3,
                                        class_mode='binary')
```

```python
train_dataset.class_indices
#train_dataset.classes prints all classes of images in o and 1's
```

```python
from tensorflow.keras import datasets, layers, models
#Create the convolutional base
#As input, a CNN takes tensors of shape (image_height, image_width, color_channels)
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
#Adding dense layers on top
#model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(64, activation='sigmoid'))
model.add(layers.Dense(10))
model.add(layers.Flatten())

#Again check architecture of model
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
from PIL import Image, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
import os
model_fit=model.fit(train_dataset,epochs= 50,steps_per_epoch= 500 ,validation_data=validation_dataset)
```

```python
print('Test accuracy:', model_fit)
```
